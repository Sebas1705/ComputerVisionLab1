## Prctica 1 - Deteccin de paneles de sealizacin vertical azul en autovas

Se desea construir un sistema para la deteccin automtica de paneles informativos de trfico de color azul, en imgenes realistas que han sido tomadas desde un coche (ver figura adjunta). Los paneles se han diseado para que sean fcilmente distinguibles del entorno en cualquier condicin de luz, color de fondo y climatologa.



Observando la imagen adjunta podemos hacernos una idea sobre el tipo de informacin que nos permitir distinguir un tipo de otro de panel, e incluso localizarlos en una imagen de carretera. En particular, se distinguen por su forma geomtrica rectangular, por las zonas en las que se divide (borde claro y fondo oscuro) o por el color del mismo (azul y blanco). Es decir, para detectar este tipo de paneles podramos utilizar:
- Un detector de regiones de alto contraste (detectaramos la parte interna del panel).
- Un algoritmo que detectase rectngulos (p.ej., combinando deteccin de lneas con detectores de esquinas).
- Un algoritmo para descubrir qu pxeles son de color azul en la imagen junto con su distribucin espacial.
- Detectar ciertas formas que aparecen en los paneles en localizaciones especiales (smbolo de salida de la autopista por la izquierda o derecha, diversos tipos de flechas, etc.).
- Cualquier otra tcnica que tenga en cuenta color y forma.

El objetivo de esta prctica es desarrollar un detector bsico de sub-paneles de informacin en autovas (como se ve en la figura adjunta). Hablamos de sub-paneles porque la idea es detectar cada una de las regiones rectangulares azules enmarcadas con un borde blanco en los paneles de la carretera. Para desarrollar el algoritmo de deteccin, como en muchos otros problemas de este tipo, se ofrece una
coleccin de imgenes de test tomadas desde un coche.
